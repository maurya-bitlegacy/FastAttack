```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

figure(figsize=(8, 6), dpi=80)
```

```python
samples=[5,10,15,20,25]
deep=[133.8,217.4,229.4,260.6,216.08]
bugger=[202.2,243.5,330.87,359.15,338.6]
bae=[184.4,238.8,298.6,296.55,287.24]
textfooler=[192,313.5,452.2,502.8,480.92]
fasttext=[153.2,219.9,239.4,256.35,242.28]

plt.figure(figsize=(5,5), dpi=100)
plt.scatter(samples,deep)
plt.scatter(samples,bugger)
plt.scatter(samples,bae)
plt.scatter(samples,textfooler)
plt.scatter(samples,fasttext)
plt.plot(samples,deep,label='DeepWordBug')
plt.plot(samples,bugger,label='TextBugger')
plt.plot(samples,bae,label='BAE')
plt.plot(samples,textfooler,label='TextFooler')
plt.plot(samples,fasttext,label='FastAttack')

plt.xlabel('Number of Samples')
plt.ylabel('Average Number of Queries')
plt.xticks(np.arange(min(samples), max(samples)+1, 5.0))
plt.legend()
plt.plot()
```

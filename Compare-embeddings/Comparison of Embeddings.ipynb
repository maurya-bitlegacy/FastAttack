```python
import numpy as np
import matplotlib.pyplot as plt
data = [[92, 52, 24,60],
[56, 8, 8,20],
[72,12,12,20]]
X = np.arange(4)
fig = plt.figure(figsize=(5,5), dpi=100)
ax = fig.add_axes([0,1,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25, label='FastText')
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25, label='Word2Vec')
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25, label='Glove')
ax.set_ylabel('Attack Success Rate')
ax.set_xticks(X)
ax.set_xticklabels(['Albert','Bert-uncased','DistilBert-uncased','Roberta']) 
ax.legend()
ax.set_title('FastText vs Word2Vec vs Glove Attack Success Rate on AG News, Greedy Strategy')
```
